```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('data/drone_dataset.pkl', 'rb') as f:
#with open('data/drone_dataset.pkl', 'rb') as f:
    my_dataset = pickle.load(f)
```

```python
# Reduce the Dataset to only 2000, instead of 4000
MAX_EPISODE_LEN = 2000
```

```python
for data in my_dataset:
    
    if len(data['observations']) > MAX_EPISODE_LEN:
        
        print("Reducing dataset to {} steps".format(MAX_EPISODE_LEN))
        data['observations'] =      data['observations'][:MAX_EPISODE_LEN]
        data['actions'] =           data['actions'][:MAX_EPISODE_LEN]
        data['rewards'] =           data['rewards'][:MAX_EPISODE_LEN]
        data['next_observations'] = data['next_observations'][:MAX_EPISODE_LEN]
        data['terminals'] =         data['terminals'][:MAX_EPISODE_LEN]
        
        
        data['rewards'][-1] = 0.0 # 0 by now! rewards[-1] = 0*env.done_reward # np.sum(np.sum(self.map_unc_min_h - self.map_unc)): terminal state reward when reaching max steps
        data['terminals'][-1] = True
        
        
```

```python
for data in my_dataset:
    print(np.max(data['actions']))
    print(np.min(data['actions']))
    plt.plot(data['rewards'])
    plt.plot(data['terminals'])
    plt.show()
```

```python
with open('data/drone_dataset.pkl', 'wb') as handle:
    pickle.dump(my_dataset, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
